```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.histplot(df['age'], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Age Distribution')

sns.histplot(df['watch_hours'], kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Watch Hours Distribution')

sns.histplot(df['monthly_fee'], kde=True, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Monthly Fee Distribution')

sns.countplot(x='churned', data=df, ax=axes[1, 1], palette='pastel')
axes[1, 1].set_title('Churned Distribution')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 15))

sns.countplot(x='subscription_type', data=df, ax=axes[0, 0], palette='muted')
axes[0, 0].set_title('Subscription Type Count')


sns.countplot(x='gender', data=df, ax=axes[0, 1], palette='muted')
axes[0, 1].set_title('Gender Count')

sns.countplot(x='region', data=df, ax=axes[1, 0], palette='muted')
axes[1, 0].set_title('Region Count')


sns.countplot(x='device', data=df, ax=axes[1, 1], palette='muted')
axes[1, 1].set_title('Device Count')


sns.countplot(x='payment_method', data=df, ax=axes[2, 0], palette='muted')
axes[2, 0].set_title('Payment Method Count')


sns.countplot(x='favorite_genre', data=df, ax=axes[2, 1], palette='muted')
axes[2, 1].set_title('Favorite Genre Count')


plt.tight_layout()
plt.show()
```

```python



plt.figure(figsize=(12, 6))
sns.barplot(x='favorite_genre', y='avg_watch_time_per_day', data=df, palette='coolwarm')
plt.title('Average Watch Time per Day by Favorite Genre')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Function to plot churn rate
def churn_rate_plot(column):
    churn_data = df.groupby(column)['churned'].mean().reset_index()
    plt.figure(figsize=(10, 5))
    sns.barplot(x=column, y='churned', data=churn_data, palette='magma')
    plt.title(f'Churn Rate by {column.capitalize()}')
    plt.ylabel('Churn Rate')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


for col in ['gender', 'region', 'subscription_type', 'payment_method']:
    churn_rate_plot(col)
```

```python
numerical_cols = df.select_dtypes(include='number')

# Compute correlation matrix
corr_matrix = numerical_cols.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
# Correlation with churned column
churn_corr = corr_matrix['churned'].sort_values(ascending=False)
print("Correlation with churned:")
print(churn_corr)
```
